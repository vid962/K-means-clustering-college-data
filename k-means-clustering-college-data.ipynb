# K Means Cluster Creation

```python
from sklearn.cluster import KMeans
```

```python
# Creating an instance of a K Means model with 2 clusters
kmeans = KMeans(n_clusters=2)
```

```python
# Fitting the model to all the data except for the Private label
kmeans.fit(df.drop('Private',axis=1))
```

```python
# clusters center vectors
kmeans.cluster_centers_
```

# Evaluation
There is no perfect way to evaluate clustering if you don't have the labels, however in this data we do have the labels, so we take advantage of this to evaluate our clusters.  **Usually we dont have that collumn**.

** Create a new column for df called **'Cluster'**, which is a 1 for a Private school, and a 0 for a public school.**

```python
def converter(cluster):
    if cluster=='Yes':
        return 1
    else:
        return 0
```

```python
df['Cluster'] = df['Private'].apply(converter)
```
